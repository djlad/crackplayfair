```python
def getRowColumn(key, letter):
    index = key.index(letter)
    return ([int(index/5), index % 5])

getRowColumn(key, "d")
```

```python
def getLetter(key, row, column):
    #gets letter from row and column
    return key[row*5 + column]

print getLetter(key, 1, 1)#r
print getLetter(key, 2, 1)#c
print getLetter(key, 1,3)#x
```

```python
def encryptDigram(key, let1, let2, decrypt = False):
    rowCol1 = getRowColumn(key, let1)
    rowCol2 = getRowColumn(key, let2)
    offset = 1
    if decrypt:
        offset = -1
    
    if rowCol1[1] == rowCol2[1]:
        return getLetter(key, (rowCol1[0]+offset)%5, rowCol1[1]) + getLetter(key, (rowCol2[0]+offset)%5, rowCol2[1])
    
    elif rowCol1[0] == rowCol2[0]:
        return getLetter(key, rowCol1[0], (rowCol1[1]+offset)%5) + getLetter(key, rowCol2[0], (rowCol2[1]+offset)%5)
        
    
    return getLetter(key, rowCol1[0], rowCol2[1]) + getLetter(key, rowCol2[0], rowCol1[1])

print encryptDigram(key, "l", "g")#yc
print encryptDigram(key, "d", "e")#od
print encryptDigram(key, "x", "g")#gq
print encryptDigram(key, "r", "n")#cu
print encryptDigram(key, "e", "x")#xm
```
